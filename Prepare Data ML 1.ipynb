#### 17. Now impute the missing values in longitudinal data which were created by combing the samples

```python
# Now after combine the methylation data, missing values are created
# We need to impute the missing values for methylation data
# We will first split the data into two, for example we have to replace columns 
# such as gender, wave etc, because they are same for both (existed and the one created)
ext_cols = ['Row.names', 'Gender', 'Wave', 'RESP', 'Sample_Name']
df1 = w1w2_comb.iloc[:, w1w2_comb.columns.str.contains('|'.join(ext_cols))]
df2 = w1w2_comb.iloc[:, ~w1w2_comb.columns.str.contains('|'.join(ext_cols))]
print("df1 shape :", df1.shape)
print("df2 shape :", df2.shape)
df1.head
```

```python
# First impute values from other columsn wherever we can for e.g sample_name etc
for i in range(len(df1.columns)//2):
    df1.iloc[:, i].fillna(df1.iloc[:, 5+i], inplace = True)
    df1.iloc[:, 5+i].fillna(df1.iloc[:, i], inplace = True)
```

```python
print("Missing values : ", df1.isna().sum().sum())
df1_filled = df1
```
